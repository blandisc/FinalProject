```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, user_agent="Mozilla/5.0 (iPhone; U; CPU like Mac OS X; en)")
```

```python
url = "https://mobile.twitter.com/search?q=%28to%3Alopezobrador_%29+until%3A2020-09-01+since%3A2020-08-01+-filter%3Alinks&s=typd&x=0&y=0"

browser.visit(url)

browser.html

soup = bs(browser.html, 'html.parser')

tweets_responses =[]

tweets_body = soup.find_all("tbody")

tweets_body1 = tweets_body[6:]

for tweet in tweets_body1:

        tweetContent = tweet.find_all('div', class_='dir-ltr')[0].text.strip()

        tweetDate = tweet.find_all('td', class_='timestamp')[0].a.text
        
        tweetUser = tweet.find_all('div', class_='username')[0].text.strip()
        
        tweets_responses.append({"Content" : tweetContent, "Date" : tweetDate, "User": tweetUser })

while True:
    try:
        browser.find_by_text('Cargar Tweets antiguos ').click()
        
        browser.html

        soup = bs(browser.html, 'html.parser')

        tweets_body = soup.find_all("tbody")

        tweets_body1 = tweets_body[6:]

        for tweet in tweets_body1:

            tweetContent = tweet.find_all('div', class_='dir-ltr')[0].text.strip()

            tweetDate = tweet.find_all('td', class_='timestamp')[0].a.text
        
            tweetUser = tweet.find_all('div', class_='username')[0].text.strip()
        
            tweets_responses.append({"Content" : tweetContent, "Date" : tweetDate, "User": tweetUser })
        
    except: 
        break
        
df = pd.DataFrame(tweets_responses)
df.sort_values(by=["Date"])
```
